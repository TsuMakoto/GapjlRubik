```julia
# 20回まわす
rotates = join(map(i -> rand(["U", "L", "F", "R", "B", "D"]), 1:20))

# 元の作成
gens = map(s -> eval(Symbol(s)), String.(split(rotates, "")))
σ = @gap ()
for gen ∈ gens; σ *= gen; end

# solve!
word = string(WordElements(σ))

println("rotates: $(rotates)")
println("state: $(string(σ))")
println("word: $(word)")
```

```julia
using Pkg
Pkg.add("HTTP");
Pkg.add("FileIO")
Pkg.add("ImageMagick")
using HTTP
using FileIO
using ImageMagick
```

```julia
problem = "http://cube.crider.co.uk/visualcube.php?size=150&case=$(rotates)&fmt=svg"
HTML(String(HTTP.request("GET", problem).body))
```

```julia
# ただし、以下のコードは解に()を含むものは実現しない。。。
alg = split(word, "*")
for i ∈ 1:length(alg)
    if '-' ∈ alg[i]
        x = split(alg[i], "^")
        alg[i] = "$(x[1])^$(parse(Int, x[2]) + 4)"
    end
end
alg = join(alg)
solved = "http://cube.crider.co.uk/visualcube.php?size=150&case=$(rotates)&alg=$(alg)&fmt=svg"
HTML(String(HTTP.request("GET", solved).body))
```

```julia
alg = split(word, "*")
solve_word = ""
for i ∈ 1:length(alg)
    if '-' ∈ alg[i]
        x = split(alg[i], "^")
        alg[i] = "$(x[1])^$(parse(Int, x[2]) + 4)"
    end
    solve_word *= alg[i]
    solved = "http://cube.crider.co.uk/visualcube.php?size=150&case=$(rotates)&alg=$(solve_word)&fmt=svg"
    io = open("./img/process$(i).html", "w")
    write(io, String(HTTP.request("GET", solved).body))
    write(io, "<a href='process$(i+1).html' > 次 </a>")
    close(io)
    sleep(1)
end
```
